## Using CASE and WHEN on Spark Data Frames

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Using CASE and WHEN on Spark Data Frames").getOrCreate()
```

```python
employees = [
    (1, 'Scott', None, 1000.0, '10',
         'United States', '+1 123 456 7890', '123 45 6789'),
    (2, 'Henry', 'Ford', 1250.0, None,
         'Italy', '+39 02 26 83 01 02.', '011 45 6789'),
    (3, 'Nick', 'Junior', 750.0, ' ',
         'United Kingdom', '+44 7975 556677', '090 56 9870'),
    (4, 'Bill', '', 1500.0, None,  
         'Australia', '+44 20 7183 8750', '119 56 9870'),
]
```

```python
employeesDF = spark.createDataFrame(employees, schema='''
    employee_id INT, first_name STRING, last_name STRING, 
    salary DOUBLE, bonus STRING, nationality STRING, contact_no STRING, ssn STRING
    ''')
```

```python
from pyspark.sql.functions import expr
```

```python
employeesDF.withColumn(
    'bonus',
    expr("""
    CASE 
        WHEN bonus IS NULL OR bonus = '' OR bonus = ' ' THEN 0
        ELSE bonus
    END
         """) \
).show()
```

```python
from pyspark.sql.functions import when, lit, isnull, col
```

```python
employeesDF.withColumn(
    'bonus',
        when((col('bonus').isNull()) | (col('bonus') == lit('')) | (col('bonus') == lit(' ')), 0).
        #when().
        #when().
        otherwise(col('bonus'))
    ) \
    .show()
```
