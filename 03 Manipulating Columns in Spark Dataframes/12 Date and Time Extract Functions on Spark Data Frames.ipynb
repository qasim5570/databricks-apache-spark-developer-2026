```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Date and Time Extract Functions on Spark Data Frames").getOrCreate()
```

```python
l = [('X',), ]
```

```python
l
```

```python
df = spark.createDataFrame(l).toDF('dummy')

df
```

```python
df.show()
```

```python
from pyspark.sql import functions as f
```

```python
from pyspark.sql.functions import current_date, current_timestamp
```

```python
df.select(
    current_timestamp().alias('current_timestamp'),
    f.day(current_timestamp()).alias('day'),
    f.month(current_timestamp()).alias('month'),
    f.year(current_timestamp()).alias('year'),
    f.dayofyear(current_timestamp()).alias('dayofyear'),
    f.dayofmonth(current_timestamp()).alias('dayofmonth'),
    f.dayofweek(current_timestamp()).alias('dayofweek'),
    f.hour(current_timestamp()).alias('hour'),
    f.minute(current_timestamp()).alias('minute'),
    f.second(current_timestamp()).alias('second'),
).show(truncate=False)
```
