```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Date and Time Arithmetic using Spark Data Frames").getOrCreate()
```

```python
from datetime import date
```

```python
list = [
    ("2023-10-26", "2023-10-26 15:30:00"),
    ("2024-02-01", "2024-02-01 10:10:10"),
    ("2023-12-31", "2023-12-31 23:59:59"),
    ("2024-03-15", "2024-03-15 09:00:00"),  # Today's date with a random time
]

list
```

```python
help(spark.createDataFrame)
```

```python
df = spark.createDataFrame(list, schema = '''
    date STRING, timestamp STRING
    ''')
```

```python
df.show()
```

```python
from pyspark.sql.functions import date_add, date_sub 
```

```python
df.withColumn('date_added', date_add(df.date, 10)) \
    .withColumn('date_subtracted', date_sub(df.date, 10)) \
    .show()
```

```python
df.withColumn('date_added', date_add(df.date, 10)) \
    .withColumn('date_subtracted', date_sub(df.date, 10)) \
    .withColumn('timestamp_added', date_add(df.timestamp, 10)) \
    .show()
```

```python
from pyspark.sql.functions import date_diff, current_date, current_timestamp
```

```python
help(date_diff)
```

```python
df.withColumn('date_difference', date_diff(current_date(), df.date)) \
    .show()
```

```python
df.withColumn('date_difference', date_diff(current_date(), df.date)) \
    .collect()
```

```python
help(df.collect)
```

```python
from pyspark.sql.functions import months_between, add_months, round
```

```python
df.withColumn('months_difference', round(months_between(current_date(), df.date), 2)) \
    .show()
```

```python
df.withColumn('months_added', add_months(df.date, 2)) \
    .withColumn('months_added_ts', add_months(df.timestamp, 2)) \
    .show()
```

```python
from pyspark.sql.functions import trunc, date_trunc
```

```python
df.select('*') \
    .withColumn('truncate_date', trunc(df.date, 'mm')) \
    .withColumn('truncate_ts', trunc(df.timestamp, 'yy')) \
    .show()
```

```python
df.select('*') \
    .withColumn('truncate_date', date_trunc('yy', df.date)) \
    .withColumn('truncate_ts', date_trunc('mm', df.timestamp)) \
    .withColumn('truncate_ts_ss', date_trunc('hour', df.timestamp)) \
    .show()
```
