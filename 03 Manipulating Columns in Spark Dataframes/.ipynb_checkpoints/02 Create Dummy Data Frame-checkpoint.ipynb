## Create Dummy Data Frame

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create Dummy Data Frame").getOrCreate()
```

```python
l = [('X', )]
```

```python
df = spark.createDataFrame(l, 'dummy STRING')
```

```python
df.show()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import current_date
```

```python
df.select(current_date()).show()
```

```python
df.select(current_date().alias('current_date')).show()
```

```python
employees = [
    (1, 'Scott', 'Tiger', 1000.0,
         'United States', '+1 123 456 7890', '123 45 6789'),
    (2, 'Henry', 'Ford', 1250.0,
         'Italy', '+39 02 26 83 01 02.', '011 45 6789'),
    (3, 'Nick', 'Junior', 750.0,
         'United Kingdom', '+44 7975 556677', '090 56 9870'),
    (3, 'Bill', 'Gomes', 1500.0,
         'Australia', '+44 20 7183 8750', '119 56 9870')
]
```

```python
len(employees)
```

```python
employeeDF = spark.\
    createDataFrame(employees,
                    schema='''employee_id INT, first_name STRING, last_name STRING, salary DOUBLE,
                    country STRING, contact_no STRING, ssn STRING''')
```

```python
employeeDF.show()
```

```python
employeeDF.printSchema()
```
