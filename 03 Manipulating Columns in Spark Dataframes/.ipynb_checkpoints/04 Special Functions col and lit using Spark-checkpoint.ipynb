## Special Functions col and lit using Spark

- lit: takes constant value as an argument, returns column object
- col: takes column name or expression name as an argument, returns column object


```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Special Functions col and lit using Spark").getOrCreate()
```

- If there are no transformations on col in a fnuction -> pass them as string
- If there are transformations, convert the column into col object then apply the transformation(function)

```python
employees = [
    (1, 'Scott', 'Tiger', 1000.0,
         'United States', '+1 123 456 7890', '123 45 6789'),
    (2, 'Henry', 'Ford', 1250.0,
         'Italy', '+39 02 26 83 01 02.', '011 45 6789'),
    (3, 'Nick', 'Junior', 750.0,
         'United Kingdom', '+44 7975 556677', '090 56 9870'),
    (4, 'Bill', 'Gomes', 1500.0,
         'Australia', '+44 20 7183 8750', '119 56 9870')
]
```

```python
employees
```

```python
employeesDF = spark.createDataFrame(employees, \
                                   schema = '''employee_id INT, first_name STRING, last_name STRING, 
                                   salary DOUBLE, nationality STRING, contact STRING, ssn STRING''')
```

```python
employeesDF.printSchema()
```

```python
employeesDF.select( \
    'first_name',
    'last_name'
).show()
```

```python
employeesDF.groupBy('nationality').count().show()
```

```python
employeesDF.orderBy('employee_id').show()
```

```python
employeesDF.select('first_name', 'last_name').orderBy('employee_id').show()
```

```python
from pyspark.sql.functions import col, concat, lit
```

```python
help(col)
```

```python
type(col)
```

```python
type(col('x'))
```

```python
employeesDF.select( \
    col('first_name'),
    col('last_name')
).show()
```

```python
employeesDF.select( \
    'first_name'.upper()
).show()
```

```python
employeesDF.orderBy(col('first_name').desc()).show()
```

```python
from pyspark.sql.functions import upper
```

```python
employeesDF.orderBy(upper(employeesDF['first_name'])).show()
```

```python
employeesDF.select(upper(employeesDF['first_name']).alias('first_name')) \
    .orderBy(upper(employeesDF['first_name'])) \
    .show()
```

```python
help(employeesDF.withColumn())
```

```python
employeesDF.withColumn('first_name', upper(employeesDF['first_name'])) \
    .orderBy(upper(employeesDF['first_name'])) \
    .show()
```

```python
employeesDF.withColumn('first_name', upper(col('first_name'))) \
    .groupBy('nationality') \
    .count() \
    .show()
```

```python
from pyspark.sql.functions import Column
```

```python
help(Column)
```

```python
help(lit)
```

```python
comppute bonus amount using 20% of salary usnig both string and col()
```

```python
employeesDF.select(
    '*',
    concat(col('first_name'), lit(', '), col('last_name')).alias('full_name')
).show()
```

```python
employeesDF.select(
    '*',
    (lit(5.00) + col('salary')).alias('bonus')
).show()
```

```python
employeesDF.select(
    '*',
    (lit(5.00) + 'salary').alias('bonus')
).show()
```
