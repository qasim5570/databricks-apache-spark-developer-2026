## Overview of Select on Spark Data Frame

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Overview of Select on Spark Data Frame").getOrCreate()
```

```python
import datetime
users_list = [
    {
        'id': 1,
        'first_name': 'John',
        'last_name': 'Doe',
        'email': 'john.doe@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 54 7400113', 'PAK_Mobile': '+92 300 5638080'},
        'is_customer': True,
        'amount_paid': 123.45,
        'customer_from': datetime.date(2023, 1, 15),  # Date in YYYY-MM-DD format
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)  # Timestamp in YYYY-MM-DD HH:MM:SS format
    },
    {
        'id': 2,
        'first_name': 'Jane',
        'last_name': 'Smith',
        'email': 'jane.smith@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 99 7666666', 'PAK_Mobile': '+92 300 0000000'},
        'is_customer': True,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    },
    {
        'id': 3,
        'first_name': 'Jack',
        'last_name': 'Alpha',
        'email': 'jack.alpha@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 99 7666555'},
        'is_customer': False,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    }
    # Add more user dictionaries as needed
]
```

```python
from pyspark.sql import Row
```

```python
users_df = spark.createDataFrame(Row(**users) for users in users_list)
users_df
```

```python
users_df.show()
```

```python
users_df.dtypes
```

```python
users_df = spark.createDataFrame([Row(**users) for users in users_list])
users_df
```

```python
users_df.dtypes
```

```python
users_df[0].show()
```

```python
users_df.select('id').show()
```

```python
type(users_df.select('id'))
```

```python
# new dataframe created implicitly

users_df.select('*').show()
```

```python
# select statement can have strings column names as arguments

users_df.select('id', 'first_name', 'last_name').show()
```

```python
# select statement can have array as an argument

users_df.select(['id', 'first_name', 'last_name']).show()
```

```python
users_df.alias('ud').select('ud.*').show()
```

```python
users_df.alias('ud').select(['ud.id', 'ud.first_name', 'ud.last_name']).show()
```

```python
from pyspark.sql.functions import col
```

```python
# col is used to use functions on top of columns of a dataframe

users_df.select(col('id'), 'first_name', 'last_name').show()
```

```python
from pyspark.sql.functions import col, concat, lit
```

```python
users_df.select \
    ('id', 
     'first_name', 
     'last_name', 
     concat(col('first_name'), lit(', '), col('last_name')).alias('full_name')) \
    .show()
```
