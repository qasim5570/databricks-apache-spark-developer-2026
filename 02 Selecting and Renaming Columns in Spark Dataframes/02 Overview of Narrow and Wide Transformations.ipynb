## Overview of Narrow and Wide Transformations

### Narrow shuffling does not result in shuffling

* Here are the functions related to narrow transformation
     * `df.select`
     * `df.filter`
     * `df.withColumn`
     * `df.withColumnRenamed`
     * `df.drop`
 
* Here are the functions related to wide transformations
     * `df.distinct`
     * `df.union`
     * `df.join` or any set operation
     * `df.groupBy` or any join operation
     * `df.sort`
     * `df.order by`
