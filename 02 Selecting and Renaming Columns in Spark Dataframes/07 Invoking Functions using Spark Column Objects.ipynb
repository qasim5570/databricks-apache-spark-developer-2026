## Invoking Functions using Spark Column Objects

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Invoking Functions using Spark Column Objects").getOrCreate()

import datetime
users_list = [
    {
        'id': 1,
        'first_name': 'John',
        'last_name': 'Doe',
        'email': 'john.doe@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 54 7400113', 'PAK_Mobile': '+92 300 5638080'},
        'is_customer': True,
        'amount_paid': 123.45,
        'customer_from': datetime.date(2023, 1, 15),  # Date in YYYY-MM-DD format
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)  # Timestamp in YYYY-MM-DD HH:MM:SS format
    },
    {
        'id': 2,
        'first_name': 'Jane',
        'last_name': 'Smith',
        'email': 'jane.smith@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 99 7666666', 'PAK_Mobile': '+92 300 0000000'},
        'is_customer': True,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    },
    {
        'id': 3,
        'first_name': 'Jack',
        'last_name': 'Alpha',
        'email': 'jack.alpha@example.com',
        'phone_numer' : {'UAE_Mobile': '+971 99 7666555'},
        'is_customer': False,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    }
    # Add more user dictionaries as needed
]
```

```python
from pyspark.sql import Row

users_df = spark.createDataFrame(Row(**users) for users in users_list)
users_df
```

### Note: All the returns from pyspsark.sql.functions return column type objects only

```python
from pyspark.sql.functions import concat, col, lit
```

```python
full_name_alias = concat(col('first_name'), lit(', '), col('last_name'))
```

```python
users_df.select('id', full_name_alias).show()
```

```python
users_df.select('id', full_name_alias).dtypes
```

```python
users_df.select('id', full_name_alias).printSchema()
```
