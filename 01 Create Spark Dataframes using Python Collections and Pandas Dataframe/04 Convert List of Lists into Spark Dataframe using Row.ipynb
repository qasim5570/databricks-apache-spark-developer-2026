## Convert List of Lists into Spark Dataframe using Row

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("04 Convert List of Lists into Spark Dataframe using Row").getOrCreate()
```

```python
users_list = [[1, 'Qasim'], [2, 'Ayesha'], [3, 'Ali'], [4, 'Hamza']]
df = spark.createDataFrame(users_list, 'user_id int, user_name string')
```

```python
df.show()
```

```python
from pyspark.sql import Row
```

```python
# Converting list of list objects into list of Row objects

df_row = [Row(*users) for users in users_list]
```

```python
type(df_row)
```

```python
type(df_row[1])
```

```python
df_row
```

```python
spark.createDataFrame(df_row, 'user_id int, user_name string')
```

```python
def dummy(*args):
    print(args)
    print(len(args))
```

```python
dummy_list = ['1', 'John']
```

```python
# treats dummy_list as a single argument since *dummy_list is not passed
dummy(dummy_list)
```

```python
# treats dummy_list as a double argument since *dummy_list is passed. *dummy_list treats each element 
# in the list as the separate element
dummy(*dummy_list)
```
