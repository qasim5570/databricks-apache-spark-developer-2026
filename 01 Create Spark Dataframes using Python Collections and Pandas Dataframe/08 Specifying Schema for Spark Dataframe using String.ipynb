## Specifying Schema for Spark Dataframe using String and List

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("06 Convert List of Dicts into Spark Dataframe using Row").getOrCreate()
```

```python
import datetime
users_list = [
    (1, 'John', 'Doe', 'john.doe@example.com', True, 123.45, datetime.date(2023, 1, 15), datetime.datetime(2024, 1, 21, 20, 5, 46)),
    (2, 'Jane', 'Smith', 'jane.smith@example.com', True, 0.0, datetime.date(2022, 11, 10), datetime.datetime(2024, 1, 21, 20, 5, 46)),
    (3, 'Jack', 'Alpha', 'jack.alpha@example.com', False, None, None, datetime.datetime(2024, 1, 21, 20, 5, 46)),
]
```

```python
# TIMESTAMP can be converted back to DATE as well but not the otherway around
users_schema = '''
    id INT,
    first_name STRING,
    last_name STRING,
    email STRING,
    is_customer BOOLEAN,
    amount_paid FLOAT,
    customer_from DATE,
    last_updated_ts TIMESTAMP
'''
```

```python
help(spark.createDataFrame)
```

```python
# Even if we don't type schema=, this will still work fine since schema= is at 2nd Postition
users_df = spark.createDataFrame(users_list, schema=users_schema)
```

```python
users_df
```

```python
# While passing the schema from array, only column names should be used
users_schema_list = [
    'id',
    'first_name',
    'last_name',
    'email',
    'is_customer',
    'amount_paid',
    'customer_from',
    'last_updated_ts']
```

```python
spark.createDataFrame(users_list, schema=users_schema_list)
```
