### Convert List of Dicts into Spark Dataframe using Row

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("06 Convert List of Dicts into Spark Dataframe using Row").getOrCreate()
```

```python
user_list = [
	{'user_id': 1, 'user_name': 'Qasim'},
	{'user_id': 2, 'user_name': 'Ayesha'},
	{'user_id': 3, 'user_name': 'Ali'},
	{'user_id': 4, 'user_name': 'Hamza'}
]
```

```python
spark.createDataFrame(user_list)
```

```python
from pyspark.sql import Row
```

### Note: Converting to a dataframe using list of dicts is deprecated

```python
# list of rows with values only
users_rows = [Row(*users.values()) for users in user_list]
users_rows
```

```python
# list of rows with keys and vaues
user_rows = [Row(**users) for users in user_list]
user_rows
```
