## Creating Multi-Column Spark DataFrame from the List

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("02 Create Multi-Column Spark Dataframe using List").getOrCreate()
```

```python
ages_list = [(15), (17), (23), (41), (32)]
```

```python
# print(ages_list)

type(ages_list[1])
```

```python
ages_list = [(15, ), (17, ), (23, ), (41, ), (32, )]
```

```python
print(ages_list)
```

```python
type(ages_list[1])
```

```python
spark
```

```python
spark.createDataFrame(ages_list)
```

```python
spark.createDataFrame(ages_list, 'ages int')
```

### spark.createDataFrame(ages_list, 'int') will fail

```python
users_list = [(1, 'Qasim'), (2, 'Ayesha'), (3, 'Ali'), (4, 'Hamza')]
```

```python
spark.createDataFrame(users_list)
```

```python
# create dataframe and define schema and column names

df = spark.createDataFrame(users_list, 'user_id int, user_name string')
```

```python
df
```
