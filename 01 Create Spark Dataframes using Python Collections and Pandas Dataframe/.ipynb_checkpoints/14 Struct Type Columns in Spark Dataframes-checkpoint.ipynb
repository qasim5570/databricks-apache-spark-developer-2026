## Struct Type Columns in Spark Dataframes

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Struct Type Columns in Spark Dataframes").getOrCreate()
```

```python
from pyspark.sql import Row
```

```python
import datetime
users_list = [
    {
        'id': 1,
        'first_name': 'John',
        'last_name': 'Doe',
        'email': 'john.doe@example.com',
        'phone_numer' : Row(UAE_Mobile='+971 54 7400113', PAK_Mobile='+92 300 5638080'),
        'is_customer': True,
        'amount_paid': 123.45,
        'customer_from': datetime.date(2023, 1, 15),  # Date in YYYY-MM-DD format
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)  # Timestamp in YYYY-MM-DD HH:MM:SS format
    },
    {
        'id': 2,
        'first_name': 'Jane',
        'last_name': 'Smith',
        'email': 'jane.smith@example.com',
        'phone_numer' : Row(UAE_Mobile='+971 99 7666666', PAK_Mobile='+92 300 0000000'),
        'is_customer': True,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    },
    {
        'id': 3,
        'first_name': 'Jack',
        'last_name': 'Alpha',
        'email': 'jack.alpha@example.com',
        'phone_numer' : Row(UAE_Mobile='+971 99 7666555', PAK_Mobile='+92 300 0000000'),
        'is_customer': False,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    },
    {
        'id': 4,
        'first_name': 'Joseph',
        'last_name': 'Marques',
        'email': 'joseph.marques@example.com',
        'phone_numer' : Row(UAE_Mobile=None, PAK_Mobile=None),
        'is_customer': False,
        'amount_paid': 0.0,
        'customer_from': datetime.date(2022, 11, 10),
        'last_updated_ts': datetime.datetime(2024, 1, 21, 20, 5, 46)
    }
    # Add more user dictionaries as needed
]
```

```python
users_df = spark.createDataFrame(users_list)
```

```python
users_df
```

```python
users_df.dtypes
```

```python
users_df.columns
```

```python
users_df.printSchema()
```

```python
# each and every column values hould have a structure, only then it will work

users_df.show()
```

```python
users_df.select('id', 'phone_numer').show(truncate=False)
```

```python
users_df \
    .select('id', 'phone_numer.UAE_Mobile', 'phone_numer.PAK_Mobile') \
    .show()
```

```python
from pyspark.sql.functions import col
```

```python
users_df \
    .select('id', col('phone_numer')['UAE_Mobile'], col('phone_numer')['PAK_Mobile']) \
    .show()
```

```python
users_df \
    .select('id', 'phone_numer.*') \
    .show()
```
