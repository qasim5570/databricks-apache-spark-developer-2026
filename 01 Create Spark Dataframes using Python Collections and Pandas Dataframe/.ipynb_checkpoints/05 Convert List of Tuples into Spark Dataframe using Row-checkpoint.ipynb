### Convert List of Tuples into Spark Dataframe using Row

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Convert List of Tuples into Spark Dataframe using Row").getOrCreate()
```

```python
spark
```

```python
users_list = [(1, 'Qasim'), (2, 'Ayesha'), (3, 'Ali'), (4, 'Hamza')]
type(users_list)
```

```python
from pyspark.sql import Row
```

```python
df = spark.createDataFrame(users_list, 'user_id int, user_name string')
df.show()
```

```python
df
```

```python
users_tuples = [Row(*users) for users in users_list]
```

```python
users_tuples
```

```python
df_tuples = spark.createDataFrame(users_tuples)
df_tuples
```

```python
df_tuples.show()
```
