## Overview Row

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("03 Row Overview").getOrCreate()
```

```python
users_list = [(1, 'Qasim'), (2, 'Ayesha'), (3, 'Ali'), (4, 'Hamza')]
df = spark.createDataFrame(users_list, 'user_id int, user_name string')
```

```python
df.show()
```

```python
# returns all records as the list of row

df.collect()
```

```python
type(df.collect())
```
