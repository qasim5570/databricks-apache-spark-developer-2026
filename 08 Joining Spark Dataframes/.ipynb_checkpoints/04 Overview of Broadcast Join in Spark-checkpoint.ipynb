## Overview of Broadcast Join in Spark

1. Also known as map side or replicated join
2. Smaller data set broadcasted to all the executors
3. Smallar data set size is driven by `spark.sql.autoBroadcastJoinThreshold`
4. When smaller data set size is greater than the threshold, we can override the join by using `broadcast` function
5. If disabled, reduce sort/sort merge join will be used

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Overview of Broadcast Join").getOrCreate()
```

```python
spark.conf.get('spark.sql.autoBroadcastJoinThreshold')
```

```python
# disable broadcast join using

spark.conf.get('spark.sql.autoBroadcastJoinThreshold', '0')
```

```python
# resetting to original value

spark.conf.get('spark.sql.autoBroadcastJoinThreshold', '10485760b')
```

```python
 %run './01 Setup Data Sets to perform joins.ipynb'
```

```python
# using reduce sort/sort merge join

users_df. \
    join(course_enrollments_df, 'user_id'). \
    count()
```

```python
from pyspark.sql.functions import broadcast
```

```python
# using broadcast join even the smaller data set size is less than the threshold
# either use broadcast() or change the configuration

broadcast(users_df). \
    join(course_enrollments_df, 'user_id'). \
    count()
```
