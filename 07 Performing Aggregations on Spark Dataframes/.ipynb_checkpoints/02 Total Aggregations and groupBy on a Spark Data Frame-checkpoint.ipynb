## Total Aggregations on a Spark Data Frame

1. Perform aggregation for order_item_product_id (say 7) to calculate the total revenue
2. Perform count of orders, sum of revenue, sum of quantity 

```python
%run './01 Validate Data Sets for Aggregations using Spark.ipynb'
```

```python
order_items.dtypes
```

```python
from pyspark.sql.functions import sum, count
```

```python
order_items.filter('order_item_order_id=7') \
    .agg(sum('order_item_product_price')) \
    .show()
```

```python
order_items.filter('order_item_order_id=7').show()
```

```python
order_items.filter('order_item_order_id=7') \
    .select( \
        count('order_item_order_id').alias('order_quantity'),
        sum('order_item_subtotal').alias('order_revenue'),
        sum('order_item_product_price').alias('total_price')
    ).show()
```

```python
order_items.filter('order_item_order_id=7').show()
```
