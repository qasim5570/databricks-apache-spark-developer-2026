## Validate Data Sets for Aggregations using Spark

1. Create Schema using `StructType` for orders and order_items
2. Read csv file using the schema created in 1

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Validate Data Sets for Aggregations using Spark").getOrCreate()
```

```python
from pyspark.sql.functions import StructType
from pyspark.sql.types import *
```

```python
schema = StructType([ 
    StructField('order_id', 
                IntegerType(), True), 
    StructField('order_date', 
                TimestampType(), True), 
    StructField('order_customer_id', 
                IntegerType(), True), 
    StructField('order_status', 
                StringType(), True)
]) 
```

```python
orders = spark.read.format('csv') \
    .schema(schema) \
    .load('./retail_db/orders/part-00000')
```

```python
orders.printSchema()
```

```python
## orders.show()
```

```python
## Order_items schema

schema_order_items = StructType([ 
    StructField('order_item_id', 
                IntegerType(), True), 
    StructField('order_item_order_id', 
                IntegerType(), True),
    StructField('order_item_product_id', 
                IntegerType(), True),
    StructField('order_item_quantity', 
                ShortType(), True), 
    StructField('order_item_subtotal', 
                FloatType(), True),
    StructField('order_item_product_price', 
                FloatType(), True),
]) 
```

```python
order_items = spark.read.format('csv') \
    .schema(schema_order_items) \
    .load('./retail_db/order_items/part-00000')
```

```python
order_items.printSchema()
```
