## Compostite Sorting

```python
from pyspark.sql.functions import col
```

```python
users_df.select('id', 'first_name', 'last_name', 'courses') \
    .sort(col('last_name').desc_nulls_first(), col('first_name').desc()) \
    .show(truncate=False)
```

```python
users_df.select('id', 'first_name', 'last_name', 'courses') \
    .sort(['first_name', 'last_name'], ascending=[1,0]) \
    .show(truncate=False)
```
