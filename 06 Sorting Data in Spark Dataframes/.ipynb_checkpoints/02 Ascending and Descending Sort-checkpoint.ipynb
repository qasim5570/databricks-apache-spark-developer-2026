### Ascending Sort

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'phone_numbers') \
    .sort('first_name') \
    .show(truncate=False)
```

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'phone_numbers') \
    .sort(users_df['last_name']) \
    .show(truncate=False)
```

```python
from pyspark.sql.functions import col
```

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'phone_numbers') \
    .sort(col('phone_numbers')) \
    .show(truncate=False)
```

```python
users_df.printSchema()
```

### Descending Sort

```python
from pyspark.sql.functions import col, desc
```

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'phone_numbers') \
    .sort(col('last_name'), ascending=False) \
    .show(truncate=False)
```

```python
from pyspark.sql.functions import size
```

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'courses') \
    .sort(size('courses'), ascending=False) \
    .show(truncate=False)
```

```python
users_df.select('id', 'first_name', 'last_name', 'gender', 'current_city', 'courses') \
    .withColumn('no_of_courses', size('courses')) \
    .sort('no_of_courses', ascending=False) \
    .show(truncate=False)
```
