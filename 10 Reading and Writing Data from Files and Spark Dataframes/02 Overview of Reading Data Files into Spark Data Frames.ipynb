## Overview of Reading Data Files into Spark Data Frames

1. Reading files `json`, `csv`, `parquet`
2. `format` and `load`
3. Specifying `option` as arguments or using `option` or `options`
4. Reading compressed files
5. `spark.read.option(inferSchema=True).toDF(*cols)`
6. using `StructType([StructField('col', IntegerType)])`
7. `help(spark.read.csv)`
8. `spark.read.csv(path, sep, schema, header)` and `spark.read.format('csv').load(path_to_csv, sep, comma, header)` can have same options
9. `spark.read.csv(path, sep, schema, header)`
10. `spark.read.options(sep, schema).csv(path).toDF()`.dtypes where `options` can be of type dict or list. In such cases, pass `spark.read.options(**options)`
11. `spark.read.format('csv').load(path_to_csv)`
12. `df.inputFiles`

## Default behaviour of `spark.read.csv`

1. delimiter: comma
2. column names will be system generated
3. all the fileds will be of type `STRING`

*While for json, parquet and ORC, schema will be inferred by default and column names and data types will be inferred using metadata that will be associated using these types of files*
