```python
users_df.filter('id = 1').show()
```

```python
from pyspark.sql.functions import col
```

```python
users_df.filter(col('id') = 1).show()
```

```python
# users_df.filter(col('id') = 1).show() does not work

users_df.filter(col('id') == 1).show()
```

```python
users_df.createOrReplaceTempView('vw_users')
```

```python
spark.sql("""
        select * from vw_users where id > 1 
    """
).show()
```
